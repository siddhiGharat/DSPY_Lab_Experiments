The most effective graphs for this purpose are:

    Heatmap (Correlation Matrix): A heatmap is highly effective for determining the correlations between more than 10 variables, displaying how different air pollutants and meteorological factors are related (e.g., positive correlation between

**Q.4     In Wine quality data set can you do non graphical EDA ? Explain your answer.**

How to Perform Non-Graphical EDA on Wine Quality Data
Using Python's pandas library, you can perform the following non-graphical techniques to understand the dataset (which includes features like fixed acidity, volatile acidity, residual sugar, pH, and alcohol):

    Viewing Data Structure (.head(), .tail(), .sample()): Inspecting the first, last, or random rows gives a quick overview of the data format and values.

    Checking Data Info (.info()): This reveals the number of entries, column names, data types (e.g., float64, int64), and memory usage. It is essential for verifying if all variables are numerical.
    Checking for Missing Values (.isnull().sum()): This identifies missing data points per column, which is crucial for determining if imputation is necessary.

    Descriptive Statistics (.describe()): This provides count, mean, standard deviation, minimum, maximum, and quartile values for all numerical features. It helps in spotting outliers and understanding the scale of data.

    Correlation Matrix (.corr()): By calculating pearson correlation coefficients, you can identify strong relationships between features (e.g., density vs. alcohol) and the target variable (quality) without using a heatmap.

    Value Counts (.value_counts()): This is used to analyze the distribution of the target variable quality (0-10) to see if the classes are balanced.
