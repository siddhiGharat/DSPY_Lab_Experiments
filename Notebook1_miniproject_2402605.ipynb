## Handling Missing Values

Different strategies are applied:

- **Age**: Filled with mean value (since it is numerical)
- **Embarked**: Filled with mode (most frequent category)
- **Cabin**: Dropped due to excessive missing values

This ensures dataset completeness.

## Filtering for Extreme Temperatures

To identify rows with extreme temperatures, we will filter the dataset based on `avg_temp` values. For this analysis, 'extreme' temperatures are defined as those below 0°C or above 30°C.

```python
# Filter for extreme temperatures (e.g., below 0 or above 30 degrees Celsius)
extremes_temp_df = df[(df['avg_temp'] < 0) | (df['avg_temp'] > 30)]

display(extremes_temp_df)
```
