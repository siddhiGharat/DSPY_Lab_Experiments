**Q.1 What is need for Hypothesis Testing ?**

Main Needs / Importance of Hypothesis Testing

**Decision Making Under Uncertainty**

In real-life situations we rarely have data for the entire population.

Hypothesis testing allows researchers to make reliable decisions using sample data.

**Testing Assumptions**

It helps verify whether a claim or assumption (hypothesis) about a population parameter is correct.

Example: Testing whether the average height of students is 165 cm.

**Scientific Research**

Researchers use hypothesis testing to validate theories and research findings.

**Comparing Groups**

It helps determine whether differences between two groups (means, proportions, etc.) are statistically significant.

**Quality Control in Industries**

Companies use hypothesis testing to check whether products meet quality standards.

**Reducing Risk of Wrong Conclusions**

It provides a structured way to control Type I and Type II errors, reducing the chance of incorrect decisions.

**Q.3 What is a student t distribution?**

**Definition**

Student’s t distribution is a probability distribution used in statistics for hypothesis testing and confidence intervals when the sample size is small and the population standard deviation is unknown.

It was introduced by William Sealy Gosset, who published his work under the name “Student.”

**Key Characteristics**

Bell-shaped and symmetric around the mean, similar to the normal distribution.

More spread out (heavier tails) than the normal distribution.

Depends on degrees of freedom (df); as df increases, the distribution approaches the normal distribution.

Used mainly for small sample sizes (n < 30).

When is the t Distribution Used?

The t distribution is used when:

The sample size is small.

The population standard deviation (σ) is unknown.

Data are approximately normally distributed.

**Example**

Suppose a researcher wants to test the average weight of students from a small sample of 15 students and the population standard deviation is unknown. In this case, a t-test using Student’s t distribution is applied.

**Conclusion**

👉 The Student’s t distribution is an important statistical tool used to make reliable inferences about a population when only small sample data are available and the population variance is unknown.



**Q.4 What is level of significance ? What role t plays in hypothesis Testing**


**1. Level of Significance (α)**

Level of significance is the probability of rejecting the null hypothesis (H₀) when it is actually true.

It represents the maximum risk of making a Type I error.

It is usually denoted by α (alpha).

Common values used in hypothesis testing are:

0.05 (5%)

0.01 (1%)

Example

If the level of significance is 0.05, it means there is a 5% chance of rejecting a true null hypothesis.

Importance

It helps set the boundary between acceptance and rejection regions.

It controls the chance of making incorrect conclusions in hypothesis testing.

**2. Role of t in Hypothesis Testing**

The t-value comes from the Student’s t-distribution and is used in hypothesis testing when:

The sample size is small (n < 30).

The population standard deviation is unknown.

Role of t Statistic

Measures Difference

The t-value measures how far the sample mean differs from the population mean.

Decision Making

The calculated t-value is compared with the critical t-value from the t-table.

Accept or Reject Hypothesis

If |t calculated| > t table value, reject H₀.

If |t calculated| ≤ t table value, accept H₀.

Used in Small Sample Tests

The t-test helps make reliable conclusions with limited data.

import numpy as np
from scipy import stats

# Sample data
ages = np.array([18, 19, 20, 18, 19, 20, 21, 19, 18, 20])

# Parameters
mu0 = 19  # hypothesized mean

# Sample statistics
n = len(ages)
x_bar = np.mean(ages)
s = np.std(ages, ddof=1)  # sample standard deviation

# t-test statistic (two-tailed)
t_stat = (x_bar - mu0) / (s / np.sqrt(n))
# p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df=n-1))

print("Sample mean:", x_bar)
print("Sample std dev:", s)
print("t statistic:", t_stat)
print("p-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: Average age is not 19")
else:
    print("Fail to reject H0: Average age may be 19")
Step 5: Decision Rule

α = 0.05 (95% confidence)

If p-value < 0.05 → Reject H₀

If p-value ≥ 0.05 → Fail to reject H₀
