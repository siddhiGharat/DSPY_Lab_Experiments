**Q.1 Explain input and output characteristics for f_oneway function from stats library**

# Input Characteristics

**Samples ()**: Takes two or more 1D array-like samples (groups) as input.

**Dimensions:** Accepts multidimensional input arrays. If inputs are 2D, the test is calculated along the first axis (columns) by default.

**Sizes:** Sample sizes can differ (unbalanced designs).

**Axis:** Optional axis parameter defines the input axis along which to compute the statistic (default is 0).

**NaN Handling:** The nan_policy parameter (propagate, omit, raise) dictates how to handle missing values.

**Method:** By default, it performs a standard one-way ANOVA. As of recent versions, equal_var=False can perform Welch's ANOVA.


# Output Characteristics
**Return Type:** Returns a tuple: (F-statistic, p-value).

**F-Statistic:** A float measuring the ratio of variance between groups relative to variance within groups.

**P-Value:** A float indicating the probability that the observed group means could occur under the null hypothesis.

**Edge Cases:**
Returns (np.nan, np.nan) if an input array has 0 length or all inputs are 1 element long.

Returns (np.inf, 0) if all values within each group are identical but differ across groups.

**Warnings:** Emits ConstantInputWarning or RuntimeWarning when encountering edge cases like zero-length arrays.


**Q.2 Explain results of sm.stats.anova_lm fucntion.**


The anova_lm() function is available in statsmodels.stats library.
It is used to perform ANOVA using linear model and gives the result in table format.

It helps to check whether the mean values of groups are significantly different.

**Syntax**

import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('y ~ x', data=data).fit()
sm.stats.anova_lm(model)

**Output of anova_lm() function**


The function returns a table with following columns:

Column Name      &      	Meaning

sum_sq	Sum of Squares

df	Degrees of Freedom

F	F statistic

PR(>F)	p-value

# Explanation of each result

**1. sum_sq (Sum of Squares)**

Shows variation in data

Higher value means more difference between groups

Example meaning:

Between groups variation

Within groups variation

**2. df (Degrees of Freedom)**

Number of independent values used in calculation

Depends on number of groups and samples

Example:

df for factor

df for error

**3. F value**

Ratio of variance between groups to variance within groups

Used to test hypothesis

Rule:

Large F → groups are different

Small F → groups are similar

**4. PR(>F) (p-value)**

Probability value

Used to decide significance

Rule:

p < 0.05 → Significant difference

p > 0.05 → No significant difference

**Q.3 What role ANOVA plays in data analytics?**

 **ANOVA (Analysis of Variance)** plays an important role in data analytics because it helps to compare the mean values of multiple groups and find whether the difference between them is significant or not.

In data analytics, we often have data from different categories, and we need to check if those categories behave differently. ANOVA helps in making this decision scientifically.

# Roles of ANOVA in Data Analytics
**1. Compare multiple groups**

ANOVA is used when we need to compare more than two groups at the same time.

Example:

Sales in 3 cities

Marks of 3 classes

Production of 3 machines

Instead of doing many t-tests, ANOVA does it in one test.

**2. Check significance of difference**

ANOVA tells whether the difference between group means is:

Real difference

OR due to random chance

Rule:

p < 0.05 → Significant difference

p > 0.05 → No significant difference

**3. Helps in decision making**

In data analytics, decisions are based on data.

ANOVA helps to decide:

Which method is better

Which product performs best

Which factor affects result

Example:

Which fertilizer gives more crop

Which advertisement increases sales

**4. Used in experiments and research**

ANOVA is widely used in:

Data science

Machine learning

Business analytics

Medical research

Education analysis

It helps to analyze experimental data.


**5. Finds effect of factors on output**

ANOVA helps to check:

Effect of one factor → One Way ANOVA

Effect of two factors → Two Way ANOVA

Example:

Marks depend on study method

Crop depends on water and fertilizer

**Q.4 When to use one way ANOVA? Explain with real world case.**

**Definition**

One Way ANOVA (One Factor ANOVA) is used when we want to compare the mean values of more than two groups based on one independent variable (factor).

It checks whether the difference between group means is significant or not.

**When to use One Way ANOVA**

Use One Way ANOVA when:

There is only one independent variable

There are three or more groups

Data is numeric

We want to compare mean values

*Example conditions:*

One factor

One dependent variable

More than two samples

**Real World Example  – Students Marks**

Suppose a teacher wants to compare marks of students from 3 classes.

Class A marks
Class B marks
Class C marks

Factor → Class
Output → Marks

We want to check:

Are marks same in all classes?

Use:

One Way ANOVA

**Q.5  When to use 2 way ANOVA ? Explain with real world case.**

**Definition**

Two Way ANOVA is used when we want to study the effect of two independent variables (factors) on one dependent variable.

It helps to check:

Effect of first factor

Effect of second factor

Combined effect of both factors

**When to use Two Way ANOVA**

Use Two Way ANOVA when:

There are two independent variables

There is one dependent variable

We want to see effect of both variables

We also want to check interaction between variables


Conditions:

Two factors

One output

More than one group in each factor

# Real World Example  – Student Marks

Suppose marks depend on:

Teaching Method (Online / Offline)

Gender (Male / Female)

Factor 1 → Method
Factor 2 → Gender
Output → Marks

We check:

Does method affect marks?

Does gender affect marks?

Does both together affect marks?

Use:

Two Way ANOVA
