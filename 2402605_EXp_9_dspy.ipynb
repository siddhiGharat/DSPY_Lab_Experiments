# **Q.1 When do we go for Chi square test ?**  

Chi-square (χ²) tests are used for categorical data, i.e., data in categories (like gender, yes/no responses, product type, etc.).

We use it when we want to:

**Test goodness of fit**: Does observed data fit an expected distribution?
Example: Are dice rolls fair?

**Test independence**: Are two categorical variables related or independent?
Example: Is gender independent of voting preference?

**Test homogeneity**: Do different populations have the same distribution of a categorical variable?

 Yes, we can use Chi-square for independence.

# **Q.3 Explain role of Critical value and p-value in Chi square test.**
Critical value (from χ² table):

Depends on significance level α (commonly 0.05) and degrees of freedom.

If χ² (calculated) > χ² (critical), reject H₀.

p-value:

Probability of observing a χ² value as extreme or more extreme than observed, assuming H₀ is true.

If p-value < α → reject H₀ → variables are not independent

If p-value ≥ α → fail to reject H₀ → no evidence of dependence

✅ Both critical value and p-value serve the same decision purpose. p-value gives more precise probability.

# **Q.4 Can we use chi square test to show existence of correlation between 2 features in a dataset ?  How will you interpret test result to indicate correlation presence.**

Chi-square does not measure correlation strength; it tests association or independence.

If you reject H₀ → there is an association between features.

If you fail to reject H₀ → no evidence of association.

Interpretation example:

χ² statistic is high, p-value = 0.01 → reject H₀ → variables are associated (presence of correlation in categorical sense).

χ² statistic is low, p-value = 0.4 → fail to reject H₀ → variables appear independent.

⚠️ Important: χ² only tells existence of association, not how strong or in what direction. For strength, measures like Cramér’s V or Phi coefficient are used.
