
Array Slicing: Accessing Subarrays

Just as we can use square brackets to access individual array elements, we can also use them to access subarrays with the slice notation, marked by the colon (:) character. The NumPy slicing syntax follows that of the standard Python list; to access a slice of an array x, use this:

x[start:stop:step]

If any of these are unspecified, they default to the values start=0, stop=size of dimension, step=1. We'll take a look at accessing sub-arrays in one dimension and in multiple dimensions.

**One-dimensional subarrays**

```
x = np.arange(10)
x
```

```
x[:5]  # first five elements
```

```
x[:5]  # elements after index five
```

```
x[4:7]  # sub-array of index 4, 5, 6
```

```
x[::2]  # every other element
```

```
x[2::2]  # every other element starting at index 2
```

A potentially confusing case is when the step value is negative. In this case, the defaults for start and stop are swapped. This becomes a convenient way to reverse an array:

```
x[::-1]  # all elements, reversed
```

```
x[3::-2]  # reversed every other from index 3
```

**Introducing UFuncs**

For many types of operations, NumPy provides a convenient interface into just this kind of statically typed, compiled routine. This is known as a vectorized operation. This can be accomplished by simply performing an operation on the array, which will then be applied to each element. This vectorized approach is designed to push the loop into the compiled layer that underlies NumPy, leading to much faster execution.


```


#Consider following loop based implementation to ind reciprocals for each element of an array
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)

```

```
#Same operatoin using UFuncs applying '/' over array elements
%timeit (1.0 / big_array)
```



Vectorized operations in NumPy are implemented via ufuncs, whose main purpose is to quickly execute repeated operations on values in NumPy arrays. UFuncs are extremely flexible – before we saw an operation between a scalar and an array, but we can also operate between two arrays as well as multidimentional arrays.

Computations using vectorization through ufuncs are nearly always more efficient than their counterpart implemented using Python loops, especially as the arrays grow in size. Any time you see such a loop in a Python script, you should consider whether it can be replaced with a vectorized expression.

**Exploring NumPy's UFuncs**

Ufuncs exist in two flavors: unary ufuncs, which operate on a single input, and binary ufuncs, which operate on two inputs.

**Q.1 What are UFuncs in numpy?**

NumPy
ufuncs (universal functions) are highly optimized, compiled C functions that perform element-wise operations on NumPy arrays. They are a core feature of NumPy used to implement vectorization, which is significantly faster than using standard Python loops, especially for large datasets.
Key Characteristics

    Vectorization: Ufuncs apply operations to entire arrays at once, pushing the loops into the underlying, efficient C code.

    Element-wise operation: A ufunc acts on corresponding elements of input arrays to produce a new output array.

    Broadcasting: Ufuncs automatically support broadcasting, a mechanism that allows them to operate on arrays of different shapes and sizes.
    
    Type Casting: They handle automatic data type conversion internally, or allow users to specify the output data type using the dtype argument.

Common Ufuncs and Methods
NumPy includes over 60 built-in ufuncs covering a wide variety of mathematical, logical, and bitwise operations. These can be broadly categorized as:

    Unary Ufuncs: Operate on a single input array (e.g., np.abs, np.sin, np.sqrt).
    Binary Ufuncs: Operate on two input arrays (e.g., np.add, np.multiply, np.maximum).

Infix operators (like +, *, /) are actually wrappers for these underlying ufuncs (e.g., a + b calls np.add(a, b)).
Ufuncs also provide powerful methods for array reduction and accumulation:

    reduce: Repeatedly applies a given operation along a specified axis of an array until a single result remains (e.g., np.add.reduce(x) computes the sum of elements in x, which is equivalent to np.sum(x)).

    accumulate: Computes the intermediate results of a reduce operation, returning an array of the same shape as the input (e.g., a cumulative sum).

    outer: Computes the operation for all pairs of elements from two different inputs to produce an outer product.

    at: Performs unbuffered in-place operations on specific elements specified by indices.



**Q.2 Which are various attributes of numy arrays object?**

NumPy array objects (instances of the
ndarray class) have several key attributes that provide information about their structure, size, and the type of data they hold.
The most commonly used attributes include:

    ndim: The number of dimensions (axes) of the array.
    shape: A tuple of integers that indicates the size of the array along each dimension. For a 2x3 array, the shape is (2, 3).
    size: The total number of elements in the array, which is the product of the elements of the shape tuple.
    dtype: An object that describes the data type of the elements in the array (e.g., int64, float64, complex128, bool).
    itemsize: The size in bytes of each element in the array. For example, an int64 array has an itemsize of 8 bytes.
    nbytes: The total number of bytes consumed by all elements of the array. This is equivalent to size * itemsize.
    data: A buffer that contains the actual elements of the array in memory.
    strides: A tuple of bytes to step in each dimension when traversing an array, which helps determine memory layout.
    flags: An object providing information about the memory layout and properties of the array, such as whether the data is contiguous or writeable.
    base: The original object from which the array was created if it is a view (a shallow copy) of another array. If the array owns its own data, this attribute is None.
    T: The transposed view of the array, a shorthand for ndarray.transpose().

**Q.4 Consider array object named Obj with dimentions 3*4*2 and you want to convert it to 4*6 shape exlpain how will you do it ? Show this using a code cell, taking an example of 3*4*2 np array of random int between 1 and 50.**

To convert a NumPy array from shape (3 × 4 × 2) to (4 × 6), we use the reshape() method.

✅ Key Concept

Reshaping is possible only if the total number of elements remains the same.

Original shape: 3 × 4 × 2 = 24 elements

Target shape: 4 × 6 = 24 elements

Since both have 24 elements, reshaping is valid.

💻 Code Example
import numpy as np

# Create 3x4x2 array with random integers between 1 and 50
Obj = np.random.randint(1, 51, size=(3, 4, 2))

print("Original Array Shape:", Obj.shape)
print(Obj)

# Reshape to 4x6
reshaped_Obj = Obj.reshape(4, 6)

print("\nReshaped Array Shape:", reshaped_Obj.shape)
print(reshaped_Obj)

🔎 What Happens Internally?

NumPy stores array data in contiguous memory (row-major order by default).

reshape() does not change the data, only how it is viewed.

The elements are reorganized logically without changing their order.

Alternative (Using -1)

You can also let NumPy calculate one dimension automatically:

reshaped_Obj = Obj.reshape(4, -1)

Here, -1 tells NumPy to compute the appropriate size automatically (which becomes 6).

 Important Note

If the total number of elements does not match, NumPy will raise:

ValueError: cannot reshape array of size X into shape (...)
