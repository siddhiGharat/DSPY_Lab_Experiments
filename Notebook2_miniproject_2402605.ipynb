## GRAPHICAL EDA
### Analysis 3: Age Distribution (Graphical)

- **Type**: Univariate Graphical Analysis  
- **Objective**: To visualize distribution of Age  

### Observation:
- Distribution is slightly right-skewed
- Majority of passengers are between 20–40 years

### Conclusion:
The dataset contains a higher proportion of young adults.

```python
plt.hist(df['age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Analysis 4: Survival vs Gender (Graphical)

- **Type**: Multivariate Graphical Analysis  
- **Objective**: To visually compare survival distribution by gender  

### Observation:
- Females show higher survival count
- Males show higher non-survival count

### Conclusion:
Visualization confirms that females had better survival chances compared to males.

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

### Analysis 5: Passenger Class vs Survival

- **Type**: Multivariate Graphical Analysis  
- **Objective**: To analyze how passenger class affects survival  

### Observation:
- First class passengers have higher survival rate
- Third class passengers have lower survival rate

### Conclusion:
Socio-economic status (Pclass) significantly influenced survival chances.

### Analysis 6: Fare vs Survival

- **Type**: Multivariate Graphical Analysis  
- **Objective**: To analyze relationship between Fare and Survival  

### Observation:
- Higher fare passengers tend to survive more
- Lower fare passengers show higher death rate

### Conclusion:
Passengers who paid higher fares (likely higher class) had better survival chances.

```python
sns.boxplot(x='survived', y='fare', data=df)
plt.title("Fare vs Survival")
plt.show()
```

## Final Conclusion

- Gender is a strong factor influencing survival
- Passenger class plays a major role in survival probability
- Age distribution shows majority are young adults
- Higher fare correlates with higher survival

Overall, socio-economic and demographic factors significantly influenced survival outcomes in the Titanic dataset.
