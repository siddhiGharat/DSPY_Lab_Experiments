**Q.1 What is 2 tailed and  tailed T Test ?**

A t-test is a statistical test used to check whether the difference between means is statistically significant. It is based on the Student’s t‑test and uses the t-distribution.

There are mainly two types based on the direction of the hypothesis:

**1️⃣ One-Tailed t-Test**

A one-tailed t-test checks if a parameter is either greater than OR less than a specific value (only one direction).

Example

Suppose a company claims the average battery life is 10 hours.

Null Hypothesis (H₀): Mean = 10

Alternative Hypothesis (H₁): Mean > 10 (right-tailed)
OR

Alternative Hypothesis (H₁): Mean < 10 (left-tailed)

Here we test only one side of the distribution.

Types of One-Tailed Tests

Right-tailed test: testing if mean > value

Left-tailed test: testing if mean < value

📊 Graphically, the rejection region is on only one side of the distribution.


**2️⃣ Two-Tailed t-Test**

A two-tailed t-test checks if the parameter is different from a specific value (either higher OR lower).

Example

Battery claim again:

H₀: Mean = 10

H₁: Mean ≠ 10

This tests both directions:

Mean could be greater than 10

Mean could be less than 10

📊 Graphically, the rejection regions are on both ends of the distribution.


**Q.2 Explain what is Type I and Type II errors with an example**

In hypothesis testing (a concept in Statistics), Type I and Type II errors describe two kinds of mistakes that can occur when deciding whether to reject a null hypothesis (H₀).

**1. Type I Error (False Positive)**

Definition:
A Type I error occurs when we reject the null hypothesis (H₀) even though it is actually true.

Simple idea:
We conclude that something exists or changed, but in reality it did not.

Example:
Suppose a medical test is checking if a person has COVID-19.

Null hypothesis (H₀): The person does not have COVID-19.

Type I error: The test result says the person has COVID-19, but the person is actually healthy.

So the test gives a false positive.

Key point:
The probability of a Type I error is called α (alpha), often set to 0.05 (5%).

**2. Type II Error (False Negative)**

Definition:
A Type II error occurs when we fail to reject the null hypothesis even though it is false.

Simple idea:
We conclude nothing changed or nothing exists, but actually it does.

Example:
Using the same COVID-19 test:

Null hypothesis (H₀): The person does not have COVID-19.

Type II error: The test says the person is healthy, but the person actually has COVID-19.

So the test gives a false negative.

Key point:
The probability of a Type II error is called β (beta).

**Q.3 What is power of statistical test?**


The power of a statistical test is a concept in hypothesis testing that measures the ability of the test to detect an effect when it truly exists. More formally:

Definition:

The power of a statistical test is the probability that the test correctly rejects the null hypothesis H0 when the alternative hypothesis H1 is true.

Mathematically:

Power=𝑃(Reject 𝐻0∣𝐻1 is true)=1-𝛽
Power=P(Reject H0∣H1is true)=1−β


**Key Points:**

High power is desirable: it means the test is good at detecting true effects.

Power depends on several factors:

Effect size: Bigger differences are easier to detect → higher power.

Sample size (n): Larger samples reduce variability → higher power.

Significance level (
𝛼
α): Higher
𝛼
α (e.g., 0.05 instead of 0.01) increases power.

Variability in data: Less variability (smaller standard deviation) → higher power.

**Example:**

Suppose you are testing whether a new drug lowers blood pressure compared to a placebo.

Null hypothesis 𝐻0H0: No difference in blood pressure.

Alternative hypothesis 𝐻1H1: Drug lowers blood pressure.

If the power of the test is 0.9, it means there is a 90% chance that the test will detect the blood pressure reduction if the drug truly works

**Q.4 Explain input and out put characteristics for  stats.ttest_rel function in your own words**


# **Input Characteristics**

stats.ttest_rel(a, b, axis=0, nan_policy='propagate')



a – First set of observations.

Usually a 1D array (like a list of numbers) or a multi-dimensional array.

Represents measurements before treatment, or from the first condition.



b – Second set of observations.

Same size as a.

Represents measurements after treatment, or from the second condition.

axis (optional) – Which axis of the array to perform the test along (default is 0).

For 1D arrays, you don’t need to change this.

For 2D arrays, it allows testing column-wise or row-wise.



nan_policy (optional) – How to handle missing values (NaNs).


'propagate' → result will be NaN if there’s a NaN in data.

'omit' → ignore NaNs in calculations.

'raise' → throw an error if NaNs are present.

**Important**: The lengths of a and b must be the same because each element in a is paired with an element in b.

# **Output Characteristics**

The function returns a tuple:

t_statistic, p_value = stats.ttest_rel(a, b)

t_statistic – The computed t-value.

Measures the difference between the paired means relative to the variation.

Large positive or negative values indicate a stronger difference.

p_value – The probability of observing such a difference if the null hypothesis is true.

Small p-value (usually < 0.05) → reject the null hypothesis → the means are significantly different.

Large p-value → fail to reject the null → no significant difference detected.

**Q.5 What is function in stats module for identifying power of a statistical test. Explain out put range of it and its significance.**

In Python, the statsmodels library (not SciPy directly) provides functions to calculate the power of a statistical test. One of the main functions is:

**Function for Power Calculation**

statsmodels.stats.power.tt_solve_power() – used for t-tests.
Other similar functions exist for z-tests, F-tests, chi-square tests, etc.

**General syntax for a t-test power calculation:**

from statsmodels.stats.power import TTestPower

power_analysis = TTestPower()

power = power_analysis.solve_power(effect_size=0.5, nobs=20, alpha=0.05, alternative='two-sided')

Input parameters explained

effect_size – Difference between means divided by standard deviation (Cohen’s d).

nobs – Number of observations in the sample.

alpha – Significance level (commonly 0.05).

alternative – Type of test: 'two-sided', 'larger', or 'smaller'.

**Output**

The output is a single value:

power – A number between 0 and 1 (sometimes expressed as %).

Range and meaning:

Value of Power	Interpretation

0.0 – 0.2	Very low probability of detecting an effect; test is weak

0.2 – 0.5	Low power; may miss true effect (high Type II error)

0.5 – 0.8	Moderate power; acceptable but could be improved

0.8 – 1.0	High power; test is reliable in detecting true effects

1.0	Perfect power; almost never misses true effects (rare in real life)

**Significance of the power value**

Helps plan experiments: Determines required sample size to detect an effect.
